# INTRODUCTION THE MACHINE LEARNING MIDTERM PROJECT

```python
#KÜTÜPHANELERİ TANIMLIYORUM
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
```

# VERİ DÜZENLEME İŞLEMLERİ

```python
#VERİLERİMİ YÜKLÜYORUM VE KONTROL EDİYORUM
veri = pd.read_csv('data.csv')
veri
```

```python
veri.isna().sum()
```

```python
label = pd.read_csv('labels.csv')
label
```

```python
#X VE Y KOLONLARIMI BELİRLİYORUM
X = veri.iloc[:, 1:]
X
```

```python
Y= label.iloc[:, 1:]
Y
```

```python
#disease_type YANİ Y KOLONUMU SAYISALLAŞTIRIYORUM 
Y = preprocessing.LabelEncoder().fit_transform(Y)
Y
```

# RANDOM FOREST MODEL EĞİTİMİ VE DEĞERLENDİRMESİ

```python
#GEREKLİ KÜTÜPHANELERİ TANIMLADIM
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix
```

```python
#CROSS VALIDATION İLE TAHMİNLEME YAPTIM
model1 = RandomForestClassifier(n_estimators=100, random_state=0) 
y_pred1 = cross_val_predict(model1, X, Y, cv=5)
```

```python
#DEĞERLENDİRME İÇİN CM'E BAKIYORUM
cm1 = confusion_matrix(Y, y_pred1)
num_classes = len(cm1)
```

```python
#HER SINIFIN SENSITIVITY VE SPECIFICITY DEĞERLERİNİ HESAPLAMAK İÇİN DÖNGÜ
for i in range(num_classes):
    TP = cm1[i, i]
    FN = sum(cm1[i, :]) - TP
    FP = sum(cm1[:, i]) - TP
    TN = np.sum(cm1) - TP - FN - FP

    sensitivity1 = TP / (TP + FN)
    specificity1 = TN / (TN + FP)

    print(f"RF Sınıf {i+1} Hassasiyet (Sensitivity):", sensitivity1)
    print(f"RF Sınıf {i+1} Özgüllük (Specificity):", specificity1)
```

```python
# SINIFLANDIRMA RAPORUNU ÇIKARDIM
report1 = classification_report(Y, y_pred1)
print(report1)
```
