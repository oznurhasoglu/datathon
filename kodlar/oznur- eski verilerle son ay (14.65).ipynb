# VERİ HAZIRLAMA

```python
import pandas as pd
import numpy as np
```

```python
veriler = pd.read_csv('train.csv', parse_dates=['Tarih'])
```

```python
veriler.isna().sum()
veriler.describe()
veriler.info()
veriler.head()
```

```python
veriler = veriler.rename(columns={"Dağıtılan Enerji (MWh)" : "DE"})
```

```python
veriler.head()
```

```python
veriler['Gün'] = veriler['Tarih'].dt.dayofweek
veriler["GünSırası"]=veriler["Tarih"].dt.day
veriler['Ay'] = veriler['Tarih'].dt.month
veriler['Yıl'] = veriler['Tarih'].dt.year
veriler['Saat'] = veriler['Tarih'].dt.hour
veriler['Mevsim'] = (veriler['Tarih'].dt.month % 12 + 3) // 3
veriler['Hafta Sonu'] = veriler['Gün'].apply(lambda x: 1 if x in [5, 6] else 0)
veriler['Gece'] = veriler['Saat'].apply(lambda x: 1 if x in [18,19,20,21,22,23,0,1,2,3,4,5] else 0)
```

```python
def add_quality(Saat):
    if 0<=Saat and Saat <=4: 
        return 0
    elif Saat>20:
        return 0
    elif 11<Saat and Saat<=12:
        return 0
    elif 6<Saat and Saat<=11:
        return 1
    elif 12<Saat and Saat<=14:
        return 1
    elif 5<=Saat and Saat<=6:
        return 2
    elif 14<Saat and Saat<=20:
        return 2
    
veriler["Artmış Mı"] = veriler["Saat"].apply(add_quality)
```

```python
veriler.head(24000)
#veriler.info()
```

```python
veriler = veriler.drop('Tarih', axis=1)
veriler= veriler.astype(float)
veri_yedek= veriler
veriler.head(24000)
#veriler.info()
```

# MODEL EĞİTME

```python
from sklearn.model_selection import train_test_split
def mape(y_test, y_pred):
    return np.mean(np.abs((y_test - y_pred) / y_test)) * 100
```

```python
X= veriler.drop(['DE'], axis=1)
X_train=[]
X_test=[]
for i in range(len(X)):
    if X.iloc[i]['Ay']== 7 and X.iloc[i]['Yıl']==2022:
       X_test.append(X.iloc[i])
    else:
        X_train.append(X.iloc[i])
```

```python
X_train = pd.DataFrame(X_train, columns=['Gün', 'GünSırası', 'Ay','Yıl', 'Saat', 'Mevsim','Hafta Sonu', 'Gece', 'Artmış Mı'])
X_test= pd.DataFrame(X_test, columns=['Gün', 'GünSırası', 'Ay','Yıl', 'Saat', 'Mevsim','Hafta Sonu', 'Gece', 'Artmış Mı'])
```

```python
Y_train=[]
Y_test=[]
for i in range(len(veriler)):
    if veriler.iloc[i]['Ay']== 7 and veriler.iloc[i]['Yıl']==2022:
       Y_test.append(veriler.iloc[i]['DE'])
    else:
        Y_train.append(veriler.iloc[i]['DE'])
```

```python
Y_train = pd.Series(Y_train)
Y_test= pd.Series(Y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
model2 = RandomForestRegressor(n_estimators=100, random_state=0) 
model2.fit(X_train, Y_train)
y_pred2 = model2.predict(X_test)
print('MAPE ', mape(Y_test, y_pred2))
```
